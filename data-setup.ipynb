```python
import pandas as pd

nri_data = pd.read_csv('NRI_Table_CensusTracts.csv')
nri_data = pd.DataFrame(nri_data)
nri_data.head(1)
```

### Filter 6 States

```python
nri_data_6states = nri_data[(nri_data['STATE']=='Utah') | (nri_data['STATE']=='California') | 
                    (nri_data['STATE']=='Washington') | (nri_data['STATE']=='Oregon') | 
                    (nri_data['STATE']=='Idaho') | (nri_data['STATE']=='Nevada')]
nri_data_6states.head(1)
```

```python
nri_final_data = nri_data_6states
for i in range (40,414):
    nri_final_data = nri_final_data.drop(nri_data_6states.columns[[i]], axis=1)
for i in range (440, 467):
    nri_final_data = nri_final_data.drop(nri_data_6states.columns[[i]], axis=1)

nri_final_data.head(1)
```

```python
nri_final_data.to_csv('CIVE202_Spring2026_GroupINFRA-02-05_Project5_RawData.csv')
```
