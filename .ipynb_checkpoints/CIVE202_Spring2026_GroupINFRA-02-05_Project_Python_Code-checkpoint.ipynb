#### Setup
Import data file and libraries
Format data as a data frame.

```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Setup data
nri_data = pd.read_csv('CIVE202_Spring2026_GroupINFRA-02-05_Project5_RawData.csv')
nri_data = pd.DataFrame(nri_data)
nri_data.head(1)
```

#### State vs State Comparisons
Compare volcanic activity between the six states.
Potentially look at, numbers of events, frequency, types of annual loss, etc\
Look for geographical trends.

```python
#Aggregate data

state_summary = nri_data.groupby("STATE").agg({
    "VLCN_EALPE" : "mean", # average expected annual loss (people affected)
    "VLCN_EALT":  "mean", # Expected annual loss - total(dollar) 
    "VLCN_EALS": "mean", #Expected annual loss score
    "VLCN_RISKV": "mean" # Overall Risk
    
}).reset_index()

#Graph or present in table
plt.figure(figsize = (14,10))
plt.suptitle("State by State Volcanic Activity Comparison", fontsize = 16)
## Avg Expected Annual Loss (population)
plt.subplot(2,2,1)
plt.bar(state_summary["STATE"], state_summary["VLCN_EALPE"])
plt.title("Avg Expected annual loss (people) vs State")
plt.xlabel("State")
plt.ylabel("Average Expected Annual Loss (population)")
plt.xticks(rotation=45)

##Avg expexted annual loss (dollar)
plt.subplot(2,2,2)
plt.bar(state_summary["STATE"], state_summary["VLCN_EALT"])
plt.title("Avg Expected annual loss (money) vs State")
plt.xlabel("State")
plt.ylabel("Average VLCN_EALT")
plt.xticks(rotation=45)

##Avg expected annual loss (score)
plt.subplot(2,2,3)
plt.bar(state_summary["STATE"], state_summary["VLCN_EALS"])
plt.title("Avg Expected annual loss (score) vs State")
plt.xlabel("State")
plt.ylabel("Average VLCN_EALS")
plt.xticks(rotation=45)

##Volcanic Risk Value
plt.subplot(2,2,4)
plt.bar(state_summary["STATE"], state_summary["VLCN_RISKV"])
plt.title("Avg Overall Risk vs State")
plt.xlabel("State")
plt.ylabel("Average VLCN_RISKV")
plt.xticks(rotation=45)

plt.subplots_adjust(hspace = 0.4, wspace = 0.3) #So the graphs don't overlap
plt.show()
```

#### Population Impacts
NRI defines risk with the following:\
<span style="color:red">Risk</span> = <span style="color:orange">Expected Annual Loss</span> x Community Risk Factor\
&nbsp;&nbsp; - Community Risk Factor = f(<span style="color:green">Social Vulnerability</span>/ <span style="color:purple">Community Resilience)</span>\
&nbsp;&nbsp; - <span style="color:orange">Expected Annual Loss</span> = Exposure x Annualized Frequency x Historical Loss Ratio\
\
Can compare different volcanic activity metrics to the states risk index or social vulnerability. Look for any trends.

```python
#Aggregate data

commun_risk_summary = nri_data.groupby("STATE").agg({
    "CRF_VALUE" : "mean", # mean community risk value per state
}).reset_index()

#Graph or present in table

plt.figure(figsize = (10,4))
sns.barplot(x=commun_risk_summary["STATE"], y=commun_risk_summary["CRF_VALUE"], hue=commun_risk_summary["STATE"],  palette="rocket")
plt.show()
```
